```
import requests
import urllib.parse
from bs4 import BeautifulSoup
from time import sleep
import copy
import pandas as pd
```

```

# Find Thailand https proxy from this link https://free-proxy-list.net/
# Edit IP Address and Port in proxy

proxy = {'https': 'http://118.174.233.41:41469',}
data = requests.get('https://tv.line.me/api/11624367/playlist/bupphesanniwat/554441/false/false', proxies=proxy)
data.encoding = 'utf-8-sig'

html=data.text
soup = BeautifulSoup(html, 'html.parser')
```

```
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```
play_list = soup.find_all('tooltip')
play_list
```

```
name_list = []

for name in play_list: #short hand of fn find_all
    print(name.get_text())
    name_list.append(name.get_text())
```

```
name_list
```

```
 view = soup.find_all('span',class_='view')
```

```
view_list = []

for num_view in view: #short hand of fn find_all
    print(num_view.get_text()[5:])
    view_list.append(num_view.get_text()[5:])
```

```
view_list
```

```
like = soup.find_all('span',class_='like')
```

```
like_list = []

for num_like in like: #short hand of fn find_all
    print(num_like.get_text()[4:])
    like_list.append(num_like.get_text()[4:])
```

```
like_list
```

```
df = pd.DataFrame(columns=['Name','Views','Likes'])
df['Name'] = name_list
df['Views'] = view_list
df['Likes'] = like_list
df
```
